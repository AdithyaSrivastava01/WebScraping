```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://outtalent.com/us50/"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
print(soup)
```

```python

# Find the table
table = soup.find('table', {'id': 'companies-table'})

# Initialize lists to store data
companies = []

# Iterate over table rows (excluding the header row)
for row in table.find_all('tr')[1:]:
    cells = row.find_all('td')
    
    rank = cells[0].get_text(strip=True)
    company = cells[1].get_text(strip=True)
    us_engineers = cells[2].get_text(strip=True)
    non_us_engineers = cells[3].get_text(strip=True)
    score = cells[4].get_text(strip=True)
    europe_offices = cells[5].get_text(strip=True)
#     global_offices = cells[6].get_text(strip=True)
    
    # Store the extracted data
    companies.append({
        'Rank': rank,
        'Company': company,
        'Software engineers in the US': us_engineers,
        'Software engineers outside the US': non_us_engineers,
        'Software engineering score': score,
        'Engineering offices in Europe': europe_offices,
        'Engineering offices in Canada, LatAm, Middle East, Africa, Asia Pacific': global_offices,
    })

# Print the extracted data
for company in companies:
    print(company)
```

```python
company_names = []
for i in companies:
    company_names.append(i['Company'])
```

```python
data  = {'Tech Companies':company_names}
```
